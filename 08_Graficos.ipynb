```python
x=np.linspace(start = 0, stop = 10, num = 50)
plt.plot(x, np.sin(x), label='sin curve')
plt.plot(x, np.cos(x), label='sin curve')
plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green')
plt.legend()   # for label
plt.title('Playing with Plots')  # for title
```

```python
plt.plot(x, np.sin(x), label='sin curve')
plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green')
plt.legend()   # for label
plt.title('Playing with Plots')  # for title
plt.xlim(1, 5)
plt.ylim(0, 0.5)
```

```python
random_array = np.random.randn(20)
```

```python
plt.plot(random_array,
        color='green',
        linestyle=':')  # dot style graph
```

```python
plt.plot(random_array,
        color='green',
        linestyle='--') # line style
```

```python
plt.plot(random_array,
        color='green',
        linestyle='--',
        linewidth=3)
```

```python
plt.plot(random_array,
        color='green',
        marker ='d')#type point
plt.plot()
```

```python
plt.plot(random_array,
        color='green',
        marker ='o', # diamond shape 
        markersize=10) # changing size of diamond shape
```

```python
plt.plot(random_array,
        color='green',
        marker ='x', # diamond shape 
        markersize=10, # changing size of diamond shape
        linestyle='None') # Remove the line show only diamond shape
```

```python
datanormal=np.random.normal(0,1,1000)
plt.hist(datanormal,density=True)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
```

```python
plt.boxplot(datanormal)
```

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,0.95,0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5,n), plt.xticks([])
plt.ylim(-1.25,+1.25), plt.yticks([])

# savefig('../figures/bar_ex.png', dpi=48)
plt.show()
```
