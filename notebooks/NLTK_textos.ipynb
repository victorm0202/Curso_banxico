# Descripciones generales de textos con `nltk` y preprocesamiento básico


```python
import nltk
import matplotlib
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
print(get_ngrams('Al inicio, los estudiantes abrieron su libro', 2))
```

- Densidad de Vocabulario y Diversidad Léxica (o riqueza de vocabulario):

```python
print('Densidad de vocabulario text1:', len(text1)/len(set(text1)))
print('Densidad de vocabulario text4:', len(text4)/len(set(text4)))
```

```python
print('Diversidad lexica text1:', len(set(text1))/len(text1))
print('Diversidad lexica text4:', len(set(text4))/len(text4))
```

```python
# grafico de dispersion lexica texto 4
text4.dispersion_plot(['citizens','democracy','freedom','duties','America'])
```

- Distribuciones de palabras y otras características

```python
from nltk.probability import *

fdist1=FreqDist(text1)

fdist1
```

```python
fdist1.plot(50,cumulative=False)
fdist1.plot(50,cumulative=True)
```

```python
fdist4=FreqDist(text4)

fdist1
```

```python
fdist4.plot(50,cumulative=True)
```

```python
# longitud de palabras
fdistW1 = FreqDist([len(w) for w in text1])
fdistW1
```

```python
fdistW1.keys()
```

```python
fdistW1.plot()
```
